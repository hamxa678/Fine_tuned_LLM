```python
from langchain_google_genai import GoogleGenerativeAI
llm = GoogleGenerativeAI(model='models/text-bison-001',google_api_key=apikey, temperature=0.5)
```

```python
!pip install unstructured
```

```python
from langchain.document_loaders import UnstructuredURLLoader
loader = UnstructuredURLLoader(
    urls=[
        'https://en.wikipedia.org/wiki/Large_language_model',
        'https://en.wikipedia.org/wiki/Islam'
    ]
    )
```

```python
data = loader.load()
```

```python
data
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n"," "],
    chunk_size=1000,
    chunk_overlap=200
)
docs = r_splitter.split_documents(data)
docs[0]
```

```python
# Import INSTRUCTOR

import InstructorEmbedding
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.vectorstores import FAISS

embeddings = HuggingFaceInstructEmbeddings()
vector_embd = FAISS.from_documents(docs, embeddings)
vector_embd
```

```python
retriever = vector_embd.as_retriever()
```

```python
rdocs = retriever.get_relevant_documents(query="Islam?")
```

```python
rdocs
```

```python
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    input_key="query",
    return_source_documents=True
)
```

```python
chain("Where is khaana kaba?")
```
